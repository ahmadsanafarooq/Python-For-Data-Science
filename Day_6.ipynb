##  Lists in Python
###  What is a List?
A **list** is an ordered collection of items that can store different data types and is **mutable**, meaning you can modify it after creation.

###  List Syntax
my_list = [item1, item2, item3]

You can store integers, strings, floats, booleans, or even other lists.

###  Examples:

```python
empty_list = []
numbers = [1, 2, 3, 4, 5,10,6,18,11,15]
mixed = ["apple", 10, 3.14, True]
nested = [[1, 2], [3, 4]]
```

##  Accessing List Elements

```python
my_list = ["a", "b", "c", "d"]
print(my_list[0])
print(my_list[-1])
```

##  Modifying Lists

```python
my_list[1] = "z"
print(my_list)
```

```python
my_list
```

##  Looping Through a List

```python
for item in my_list:
    print(item)
```

###  Enumerate (with index)

```python
for i, item in enumerate(my_list):
    print(i, item)
```

##  List Comprehensions

```python
squares = [x**2 for x in range(10)]
```

```python
squares
```

With condition:

```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
```

```python
even_squares
```

##  Tuples in Python
###  What is a Tuple?
A **tuple** is similar to a list, but it is **immutable** – you cannot add, remove, or modify elements.

image.png

###  Tuple Syntax

```python
my_tuple = (item1, item2, item3)
```

###  Examples:

```python
empty_tuple = ()
single = (5,)
mixed_tuple = ("apple", 1, 3.14, False)
```

##  Accessing Tuple Elements

```python
my_tuple = ("a", "b", "c")
print(my_tuple[0])
print(my_tuple[-1])
```

##  Tuple Immutability

```python
my_tuple[1] = "z"  #  Raises TypeError
```

##  Tuple Methods

### 1. `count(value)` – Count occurrences

```python
my_tuple.count("a")
```

### 2. `index(value)` – Index of value

```python
my_tuple.index("b")
```

##  Conversion: List ↔ Tuple

### List to Tuple

```python
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
```

### Tuple to List

```python
my_tuple = (1, 2, 3)
my_list = list(my_tuple)
```

##  List vs Tuple

| Feature       | List           | Tuple           |
|---------------|----------------|------------------|
| Syntax        | `[]`           | `()`             |
| Mutability    |  Mutable     | Immutable     |
| Methods       | Many           | Few              |
| Performance   | Slower         | Faster           |
| Use Case      | Dynamic data   | Fixed structure  |

##  Intermediate List Techniques

### 1. Nested List Access

```python
matrix = [[1, 2], [3, 4]]
print(matrix[1][0])
```

```python
print(matrix[0][0])
```

### 2. Unpacking Lists and Tuples

```python
a, b, c = [1, 2, 3]
print(a)
print(b)
print(c)
```

### 3. Slicing

```python
my_list = [0, 1, 2, 3, 4]
print(my_list[1:4])
```

### 4. Check Membership

```python
"apple" in ["banana", "apple", "cherry"]
```
